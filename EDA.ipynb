```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('2015-census.csv')
print(data.head())
```

```python
print(data.info())
print(data.isnull().sum())
print(data.fillna().mean(), inplace=True)
print(data.drop_duplicates(inplace=True))
```

```python
print(data.describe())
```

```python
sns.countplot(x='health', data=data)
plt.title('Распределение состояния деревьев')
plt.show()
```

```python
data.hist(bins=30, figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
for column in data.select_dtypes(include=[np.number]).columns:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=data[column])
    plt.title(f'Коробчатая диаграмма для {column}')
    plt.show()
```

```python
plt.scatter(data['block_id'], data['tree_dbh'])
plt.title('График рассения между block_id и tree_dbh')
plt.xlabel('Column X')
plt.ylabel('Column Y')
plt.show()
```

```python
plt.scatter(data['st_assem'], data['st_senate'])
plt.title('График рассеяния между st_assem and st_senate')
plt.xlabel('Column X')
plt.ylabel('Column Y')
plt.show()
```
